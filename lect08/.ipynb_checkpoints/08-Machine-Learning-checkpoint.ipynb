## Statistical Models

Before we discuss about machine learning, we need to first talk about statistical models or _models_. What is a _model_? A *statistical model* is a summary or a brief description of the observed data. We generally want the model to summarize the data as closely as possibly but also be as simple as possible. A "good" model should also tell us about the population of the sampled data, not the sample itself. As we cannot directly measure a population, so the best we can hope for is to make "generalizations" from a sample to a population using the model.


```python
np.random.seed(98189)
n = 500
m = 20
x = np.random.randint(1,100,n)
y = x + np.random.randn(n)*20
f, axarr = plt.subplots(2, figsize=(10,16))
axarr[0].plot(x,y,'bs',alpha=0.5)
sx = np.random.randint(1,100,m)
sy = sx + np.random.randn(m)*10
axarr[0].plot(sx,sy,'ro',alpha=1.0,ms=10)
axarr[0].legend(['population','sample'])
axarr[1].plot(sx,sy,'ro',alpha=1.0,ms=10)
axarr[1].legend(['sample'])
```

More specifically, a model is a specification of a mathematical (or probabilistic) relationship that exists between one or more variables. There are a wide range of models. Some models are just simple mathematical relationships between two or more variables. Some models don't describe the relationship directly, but rather, describe the relationship in terms of probability densities. These are also known as, *probabilistic models*.

Let's take a look at a few examples of models for height and weight of a person.

```python
# Suppose we collect weight / height data from 30 people
np.random.seed(72)
n = 30
h = np.random.randint(150,180,n)
w = np.round(h - 100 + (np.random.randn(n)*5-10),1)
data = sorted(list(zip(w,h)))

# Preview 
for idx,entry in enumerate(data[:10]):
    print('Student {}: weight = {}, height = {}'.format(idx+1,entry[0],entry[1]))
```

```python
w,h = zip(*data)
plt.figure(figsize=(10,7))
plt.plot(w,h,'ro')
plt.ylabel('height')
plt.xlabel('weight')
```

From the scatter plot above, we could model the relationship between weight and height using the following a **linear model**:

$$height = weight + 110$$

Below is the plot of the model.

```python
# simple linear model
def linear_height(weight):
    return weight+110 

linear_h = np.apply_along_axis(linear_height,0,w)
plt.figure(figsize=(10,7))
plt.plot(w, h, 'ro')
plt.plot(w, linear_h, 'b-')
```

Does the linear model make sense? As you can see from the plot above, it fits the data quite well. However, if we were to use this model to predict height of a person who weighs 150kg, 
$$
height = weight + 110 = 150 + 110 = 260
$$

The predicted height is way too high. Let's consider an alternative model.

$$
height = 50\sqrt{\frac{weight}{20}} + 80
$$


```python
# second-order model
def so_height(weight):
    return ((weight/20)**0.5)*50+80

so_h = np.apply_along_axis(so_height,0,w)
plt.figure(figsize=(10,7))
plt.plot(w, h, 'ro')
plt.plot(w, so_h, 'b-')
```

```python
so_height(150)
```

The model above predicts a person who weighs 150kg would have height of 217cm, which is lower than the prediction from the linear model. 

The two models describe the relationship between $height$ and $weight$. Sometimes, we want to create a model that doesn't give a single-value prediction, but instead, a probability density function.

```python
# probabilistic model
def prob_height(weight, color='r'):
    mu = ((weight/24.0)**0.5)*100 
    sigma = 10
    x = np.linspace(mu-4*sigma, mu+4*sigma, 100)
    plt.plot(x,norm.pdf((x-mu)/sigma),'-'+color,)
    
plt.figure(figsize=(10,7))
prob_height(60,color='r')
prob_height(85,color='b')
plt.legend(['w=60','w=85'])
plt.xlabel('height')
```

## Machine Learning

Suppose we have collected some data. How do we build a model? This is an extremely important question in data science. Let's explore our options.

Traditionally, models are built/trained from domain-specific experts. For example, models in biology are likely to be carefully hand-crafted by biologists who have good insights into the collected data. For example, if someone is looking to build a model for two variables $x$ and $y$, one would need to really understand how $x$ and $y$ are related through a series of experiments and such.

However, this trend is changing. As the amount of collected data increases, the process of model training become more and more automated. This is where machine learning comes in.  

Machine learning is considered a subfield of Artificial Intelligence (AI) that specializes in techniques for training models from data, model adaptation and generalization of models. It has been applied to many other domains such as computer visions, medicines, business etc. Oftentimes, these trained models are used in systems to make decisions on their own so that they seem to be "smart". For example, Siri, a voice recognition module, that is basically a trained model that translates voice commands into user intents.

In short, machine learning is a process of creating and using models that are *learned from data*. Models trained by machine learning are commonly for the following tasks:
1. **Regression** -- Given a set of numerical features as input, predict a numerical output. For example, predicting the height of a person given the weight in the previous example.  
1. **Classification** -- Given a set of features of an object, predict its categorical label. For example, predicting whether an email is spam, or predicting whether a credit card transaction is fradulent, etc.
1. **Clustering** -- Given a data set, group the data points into clusters.


## Training a model

To demonstrate the training process, we will focus first on the **regression** task. Consider the height and weight data above. Suppose we want to train a model to **predict** height of a person using his/her weight. 

First, we need to decide on a **parameterized** family of models. For instance, we might start by choosing a linear model:

$$
height = \alpha*weight + \beta
$$

where $\alpha$ and $\beta$ are model parameters.

Next, we will use the data to **learn** the "best" value of $\alpha$ and $\beta$ to use. This step is also known as *parameter learning*, *model training* or *model fitting*.

But, what is "best"? Typically, the criteria for choosing best parameters is specific to each learning task and the nature of data. Keep in mind that the ultimate goal here is not to train a model that fits the sample well, but rather a model that generalizes well. It is common to define what best in terms of an objective function and pose the problem of finding the best parameters as an **optimization problem**.

In our case, let's define an error function given $\alpha$ and $\beta$ as follows. 

$$
error(\alpha,\beta) = \sum_{i=1}^{n} (\alpha w_i + \beta - h_i)^2
$$

where $(w_i,h_i)$ is the weight and height of example $i$ in our dataset. Now that we have a clear objective, we can find $\alpha$ and $\beta$ that minimizes the error function. Intuitively, we want to find $\alpha$ and $\beta$ so that the model predicts the height as close to the observation as possible. Note that $\alpha$ and $\beta$ is sample dependent, meaning if we have a different sample, then we will get different values of $\alpha$ and $\beta$.

Here's how you can find $\alpha$ and $\beta$ for now. Details will be covered later on.

```python
from sklearn import linear_model

# Create linear regression object
regr = linear_model.LinearRegression()

# Train the model using the training sets
regr.fit(np.array(w).reshape(-1,1), h)

# coefficient
alpha = regr.coef_
print("alpha= %.4f" % alpha)

# intercept
beta = regr.intercept_
print("beta= %.4f" % beta)

plt.figure(figsize=(10,7))
plt.plot(w,h,'ro')
x = np.linspace(min(w),max(w),100)
plt.plot(x,x*alpha+beta,'b-')
plt.xlabel('weight')
plt.ylabel('height')
```

## Making prediction


So, we have now a trained model:

$$
\hat{height} = 0.8391 \times weight + 118.1140 
$$

This model allows us to make prediction given a person's weight. Note that we usually use the $^$ notation to note that a variable is a prediction.

```python
def predict_height(weight):
    return alpha * weight + beta

print(predict_height(85))
print(predict_height(70))
```

## Overfitting and underfitting

A common problem that arises in model training is that the trained model generalizes poorly. For predictive models, this means the predictions are poor for new data (data not observed during the training). Typically, models that generalize poorly are either **overfitting** or **underfitting**. 


- **Overfitting** -- A model is overfitting is when it fits well on the data we train on but generalizes poorly to new data. 
- **Underfitting** -- A model is underfitting is when it doesn't perform well even on the training data. 


Overfitting and underfitting is closely related to the complexity of the model itself. When a model is too complex, it is more likely that it is going to be overfitting. On the other hand, when a model is too simple, it is more likely that it is going to be underfitting. 

Let's take a look at the toy example below.


```python
np.random.seed(101)
x = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
y = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
plt.figure(figsize=(12,8))
plt.plot(x,y,'sr')
plt.vlines(7.3,-2,12,color='g',linestyle='--')
p0 = np.poly1d(np.polyfit(x,y,0))
p1 = np.poly1d(np.polyfit(x,y,1))
p9 = np.poly1d(np.polyfit(x,y,9))
nx = np.linspace(-1,11,100)
plt.plot(nx,p1(nx),'-b',label="degree-1 most appropriate")
plt.plot(nx,p0(nx),'-m',label="degree-0 underfitting")
plt.plot(nx,p9(nx),'-k',label="degree-9 overfitting")
plt.legend(loc=2)

# for k in range(100):
#     x = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
#     y = np.array([i + np.random.rand(1)*1.5 for i in range(10)]).reshape((10,))
#     plt.plot(x,y,'ob',alpha=0.2)

plt.ylim([-2,12])
plt.xlim([0,12])
```

The red data points are observations. Here we compare three models:
1. polynomial degree-0 (constant) 
1. polynomial degree-1 (linear)
1. polynomial degree-9

As you can see, the training error decreases as the model complexity increases. However, consider when we want to apply this model to the predict the outcome $y$ when $x = 7.3$. In this case, the linear model probably is going to work best as the poly degree-9 is too "specialized" to the training point, and the poly degree-0 is too simple to even capture the training data.

We can say that the poly degree-9 model is **overfitting** while the poly degree-0 model is **underfitting**.




Now consider another toy example, but this time for **classification** task. Suppose each observed data point belongs to either a "red" class or a "blue" class. The goal here is to train a model to predict whether a data point is red or blue. Here's the data avaliable to use during training.

## How to deal with overfitting and underfitting

How do we know if our model is overfitting or underfitting? A fundamental way to check is to use different set of data for training the model and for testing the model.

It is common to split our data set into train and test set. For example, we use about 75% for training the model and 25% for testing. If the model was overfit to the training data, then it will perform poorly on the test data. On the other hand, if the model performs well on the test data, then we gain some confident that our model generalizes quite well.

Sometimes, you might want to train multiple models and then choose the "best" performing one. The common practice is to have another set of data called *validation set* for this purpose. To summarize:
 - **Training set** -- for training models only
 - **Validation set** -- for choosing model and/or tuning parameters
 - **Test set** -- for evaluating the final model

## In-class exercise

We will train simple models to classifier plant species from a collection of shape and texture features extracted from digital images of leaf specimens. Original dataset from https://archive.ics.uci.edu/ml/datasets/leaf.


For classification task, how do we know if a preditive model is performing well or not? For each prediction, there are total of 4 outcomes:
 - True positive (TP)
 - False positive (Type 1 Error) (FP)
 - False negative (Type 2 Error) (FN)
 - True negative (TN)
 
<img src="https://blog.exsilio.com/wp-content/uploads/2016/09/table-blog.png">
 
Typically we can evaluate performance of model using the following measurements:

 - Accuracy = (TP+TN)/(TP+FP+FN+TN)
 - Precision = TP/(TP+FP)
 - Recall = TP/(TP+FN)
 - F1-Score = 2 \* (Recall \* Precision) / (Recall + Precision)
 
 
For example, consider a predictive model that predicts the person has cancer if his name is John. Here's the prediction matrix.

<table>
<tr>
<td></td><td>John</td><td>not John</td>
</tr>
<tr>
<td>Has cancer</td><td>50</td><td>5000</td>
</tr>
<tr>
<td>No cancer</td><td>15000</td><td>950000</td>
</tr>
</table>



<div class="alert alert-block alert-success">
    <b>Your task:</b> Compute accuracy, precision, recall, and f1_score from the given prediction matrix above.
    
 </div>

```python
# Show your work
tp = 50
tn = 950000
fp = 15000
fn = 5000

def cal_accuracy(tp, tn, fp, fn):
    return (tp+tn)/(tp+tn+fp+fn)

def cal_precision(tp, fp):
    return tp/(tp+fp)

def cal_recall(tp, fn):
    return tp/(tp+fn)

def cal_f1s(tp, fn, fp):
    recall = cal_recall(tp, fn)
    precision = cal_precision(tp, fn)
    return (2*recall*precision)/(recall+precision)

print(cal_accuracy(tp, tn, fp, fn))
print(cal_precision(tp, fp))
print(cal_recall(tp, fn))
print(cal_f1s(tp, fn, fp))
```

Now, lets focus on the leaf classification task. To simplify things, we will consider a *binary* classification for now. Our goal here is to classify whether a given leaf is a class-1 or not. 

```python
from sklearn.utils import shuffle


names = [
    'Class (Species)',
    'Specimen Number', 
    'Eccentricity',
    'Aspect Ratio', 
    'Elongation',
    'Solidity',
    'Stochastic Convexity',
    'Isoperimetric Factor',
    'Maximal Indentation Depth',
    'Lobedness',
    'Average Intensity', 
    'Average Contrast', 
    'Smoothness', 
    'Third moment',
    'Uniformity', 
    'Entropy']

df = pd.read_csv("leaf.csv", header=None, names=names)

df['IsClassOne'] = df['Class (Species)'].apply(lambda x: x==1)
df.head()
```

```python
# Show your work
shuffled_df = shuffle(df)

for_train, for_test = np.split(shuffled_df, [int(.75*len(df))])
```

Consider the following simple classifer.

> If aspect ratio of a leaf > 2, then it is class 1. Otherwise, it is not class 1.

<div class="alert alert-block alert-success">
    <b>Your task:</b> Report <b>train</b> accuracy, precision, recall, and f1_score of the classifier.
 </div>

```python
# Show your work
for_train.head()
```

```python
def count_tp(df):
    is_MoreThan2_Class1 = df[(df['Aspect Ratio'] > 2) & (df['IsClassOne'])]['Class (Species)'].count()

    return is_MoreThan2_Class1

def count_tn(df):
    is_LessThan2_NotClass1 = df[(df['Aspect Ratio'] <= 2) & (df['IsClassOne'] == False)]['Class (Species)'].count()
    return is_LessThan2_NotClass1

def count_fp(df):
    is_LessThan2_Class1 = df[(df['Aspect Ratio'] <= 2) & (df['IsClassOne'])]['Class (Species)'].count()
    return is_LessThan2_Class1

def count_fn(df):
    is_MoreThan2_NotClass1 = df[(df['Aspect Ratio'] > 2) & (df['IsClassOne'] == False)]['Class (Species)'].count()
    return is_MoreThan2_NotClass1

train_tp = count_tp(for_train)
train_tn = count_tn(for_train)
train_fp = count_fp(for_train)
train_fn = count_fn(for_train)
```

```python
print(cal_accuracy(train_tp, train_tn, train_fp, train_fn))
print(cal_precision(train_tp, train_fp))
print(cal_recall(train_tp, train_fn))
print(cal_f1s(train_tp, train_fn, train_fp))
```
